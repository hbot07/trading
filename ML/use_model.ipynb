```python
from tensorflow.keras.models import load_model
```

```python
# Returns a compiled model identical to the previous one
model = load_model('predict_10m.h5')
```

```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
nifty_500df = pd.read_csv('ind_nifty500list.csv')
nifty_500df.head()
```

```python
window = 60
X_test = np.empty((0, window))  # 'window' is your defined window size, here 60
symbols = []

for symbol in nifty_500df['Symbol']:
    data = yf.download(tickers=symbol+".NS", interval='1m', period='2h')
    if data['Close'][-1] > 100:
        continue
    prices = np.array(data['Close'])
    window = 60
    X_train_symbol = np.array([prices[-60:]])

    symbols.append(symbol)    
    X_test = np.concatenate((X_test, X_train_symbol), axis=0)
```

```python
X_test.shape
```

```python
#make predictions on X_test
stock_data = pd.DataFrame()

symbols = np.array(symbols)
stock_data['Symbol'] = symbols

stock_data['Price'] = X_test[:,-1]

predictions = model.predict(X_test)
stock_data['Predictions'] = predictions

profit = predictions - X_test[:,-1].reshape(-1,1)
stock_data['Profit'] = profit

profit_percent = profit / X_test[:,-1].reshape(-1,1) * 100
stock_data['Profit Percent'] = profit_percent

stock_data.sort_values('Profit Percent', ascending=False, inplace=True)
stock_data
```
